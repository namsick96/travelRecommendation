```python
import numpy as np
import pandas as pd
import folium
```

```python
food = pd.read_csv("food_coor.csv")
alcohol = pd.read_csv("alcohol_coor.csv")
cafe = pd.read_csv("cafe_coor.csv")
tourism = pd.read_csv("tourism_coor.csv")
#food
#cafe
```

```python
long = food['x'].mean()
lat = food['y'].mean()

#map1 = folium.Map([lat,long], zoom_start = 10.4)
#map1
```

```python
map_food = folium.Map([lat,long], zoom_start = 10.4)


for i in food.index:
    sub_lat = food.loc[i,'y']
    sub_long = food.loc[i,'x']
    folium.Marker([sub_lat, sub_long], icon=folium.Icon(color = "red")).add_to(map_food)
map_food
```

```python
map_cafe = folium.Map([lat,long], zoom_start = 10.4)


for i in cafe.index:
    sub_lat = cafe.loc[i,'y']
    sub_long = cafe.loc[i,'x']
    folium.Marker([sub_lat, sub_long], icon=folium.Icon(color = "green")).add_to(map_cafe)
map_cafe
```

```python
map_alcohol = folium.Map([lat,long], zoom_start = 10.4)


for i in alcohol.index:
    sub_lat = alcohol.loc[i,'y']
    sub_long = alcohol.loc[i,'x']
    folium.Marker([sub_lat, sub_long], icon=folium.Icon(color = "orange")).add_to(map_alcohol)
map_alcohol
```

```python
map_tourism = folium.Map([lat,long], zoom_start = 10.4)


for i in tourism.index:
    sub_lat = tourism.loc[i,'y']
    sub_long = tourism.loc[i,'x']
    folium.Marker([sub_lat, sub_long], icon=folium.Icon(color = "blue")).add_to(map_tourism)
map_tourism
```

```python
category = [food, cafe, alcohol, tourism]
color = ['red','green','orange','blue']

map_total = folium.Map([lat,long], zoom_start = 10.4)

for cate, c in zip(category,color):
    for i in cate.index:
        sub_lat = cate.loc[i,'y']
        sub_long = cate.loc[i,'x']
        folium.Marker([sub_lat, sub_long], icon=folium.Icon(color = c)).add_to(map_total)

map_total
```
